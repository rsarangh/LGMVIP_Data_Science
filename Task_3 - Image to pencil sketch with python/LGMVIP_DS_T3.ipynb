## 1. Importing libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
%matplotlib inline
print("Imported libraries successfully!")
```

## Loading the Image

```python
image=cv2.imread("h.jpg")
image = cv2.resize(image,(500,700))
image
```

```python
img_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

###### Converting the Image into Grayscale

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray_image
```

```python
plt.imshow(gray_image)
```

## Inverting the Grayscaled Image

```python
inverted_image = cv2.bitwise_not(gray_image)
inverted_image
```

```python
plt.imshow(inverted_image)
```

## Blurring the Inverted Image

```python
img_blr = cv2.GaussianBlur(inverted_image, (21, 21),sigmaX=0, sigmaY=0)
img_blr
```

```python
plt.imshow(img_blr)
```

## Inverting the blurred image

```python
def dodgeV2(x, y):
    return cv2.divide(x, 255 - y, scale=256.0)
```

```python
pencil_sketch = dodgeV2(gray_image, img_blr)
plt.imshow(pencil_sketch)
```

## Comparing the Original Image with Pencil Sketch

```python
display(Image.fromarray(img_rgb))
display(Image.fromarray(pencil_sketch))
```
